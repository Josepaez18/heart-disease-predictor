# Predicción de Enfermedades Cardíacas
## Modelo: Regresión Logística
**Dataset:** Heart Disease UCI — archive.ics.uci.edu

```python
import pandas as pd
```

```python
heart = pd.read_csv('heart.csv')
```

```python
heart.head()
```

```python
# Ver cuántas filas y columnas tiene el dataset
heart.shape
```

```python
# Eliminar filas con valores nulos
heart = heart.dropna()
heart.shape
```

```python
# Separar variables de entrada (X) y variable objetivo (y)
# target: 0 = sin enfermedad, 1 = con enfermedad
X = heart.drop('target', axis=1)
y = heart['target']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)
```

```python
y_pred = logreg.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                               display_labels=['Sin Cardiopatía', 'Con Cardiopatía'])
disp.plot(cmap='Blues')
plt.title('Matriz de Confusión - Regresión Logística')
plt.show()
```

```python
from sklearn.metrics import accuracy_score, classification_report

print('Accuracy:', accuracy_score(y_test, y_pred))
print()
print(classification_report(y_test, y_pred,
      target_names=['Sin Cardiopatía', 'Con Cardiopatía']))
```
