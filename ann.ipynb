# Predicción de Enfermedades Cardíacas
## Modelo: Red Neuronal Artificial (ANN)
**Dataset:** Heart Disease UCI — archive.ics.uci.edu

```python
import pandas as pd
```

```python
heart = pd.read_csv('heart.csv')
```

```python
heart.head()
```

```python
# Ver cuántas filas y columnas tiene el dataset
heart.shape
```

```python
# Eliminar filas con valores nulos
heart = heart.dropna()
heart.shape
```

```python
# Separar variables de entrada (X) y variable objetivo (y)
# target: 0 = sin enfermedad, 1 = con enfermedad
X = heart.drop('target', axis=1)
y = heart['target']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42
)
```

```python
# StandardScaler: normaliza los datos para que todas las columnas
# tengan la misma escala. Esto es importante para la Red Neuronal
# porque si una columna tiene valores muy grandes (ej: colesterol=250)
# y otra muy pequeños (ej: sex=0 o 1), la red no aprende bien.
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)  # ajusta y transforma el train
X_test  = scaler.transform(X_test)       # solo transforma el test

print('Scaler listo')
```

```python
# Red Neuronal Artificial con 3 capas ocultas: 64, 32 y 16 neuronas
from sklearn.neural_network import MLPClassifier

ann = MLPClassifier(
    hidden_layer_sizes=(64, 32, 16),
    activation='relu',
    max_iter=500,
    random_state=42
)
ann.fit(X_train, y_train)
```

```python
y_pred = ann.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                               display_labels=['Sin Cardiopatía', 'Con Cardiopatía'])
disp.plot(cmap='Reds')
plt.title('Matriz de Confusión - Red Neuronal Artificial (ANN)')
plt.show()
```

```python
from sklearn.metrics import accuracy_score, classification_report

print('Accuracy:', accuracy_score(y_test, y_pred))
print()
print(classification_report(y_test, y_pred,
      target_names=['Sin Cardiopatía', 'Con Cardiopatía']))
```

```python
# Curva de pérdida durante el entrenamiento
plt.figure(figsize=(8, 4))
plt.plot(ann.loss_curve_, color='red')
plt.xlabel('Épocas')
plt.ylabel('Pérdida')
plt.title('Curva de Entrenamiento - ANN')
plt.grid(True, alpha=0.3)
plt.show()
```
